```python
import sys
import cv2
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
def plot_img(name, img):
    plt.imshow(img)
    plt.title(name)
    plt.show()
```

```python
im = cv2.imread("./fapiao/croped/80.14.png")
plot_img('im', im)
gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
plot_img('gray', gray)
thresh = cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV,21,11)
plot_img('thresh', thresh)
```

```python
courrier = cv2.imread("./courrier-fonts.png")
courrier_gray = cv2.cvtColor(courrier, cv2.COLOR_BGR2GRAY)
courrier_thresh = cv2.adaptiveThreshold(courrier_gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV,21,2)

one = courrier_thresh[8:58,330:363]
reduced_one = cv2.resize(one,(13,20))

two = courrier_thresh[8:58,365:398]
reduced_two = cv2.resize(two,(13,20))

three = courrier_thresh[8:58,400:433]
reduced_three = cv2.resize(three,(13,20))

four = courrier_thresh[8:58,437:470]
reduced_four = cv2.resize(four,(13,20))

five = courrier_thresh[8:58,475:508]
reduced_five = cv2.resize(five,(13,20))

six = courrier_thresh[8:58,510:543]
reduced_six = cv2.resize(six,(13,20))

seven = courrier_thresh[8:58,546:579]
reduced_seven = cv2.resize(seven,(13,20))

eight = courrier_thresh[8:58,582:615]
reduced_eight = cv2.resize(eight,(13,20))

nine = courrier_thresh[8:58,617:650]
reduced_nine = cv2.resize(nine,(13,20))

zero = courrier_thresh[8:58,295:328]
reduced_zero = cv2.resize(zero,(13,20))

digits = [reduced_zero, reduced_one, reduced_two, reduced_three, reduced_four, 
          reduced_five, reduced_six, reduced_seven, reduced_eight, reduced_nine]

for digit in digits:
    digit[np.nonzero(digit)] = 255

def detect_number(reduced, candidate):
    red = reduced
    if (candidate.shape != red.shape):
        red = cv2.resize(reduced, np.transpose(candidate).shape)
    red[np.nonzero(red)] = 255
    conv = np.multiply(red, candidate)
    return np.sum(np.sum(conv))
    #plot_img("Detect", conv)

def detect_one(candidate):
    return detect_number(reduced_one, candidate)

def detect_two(candidate):
    return detect_number(reduced_two, candidate)

def detect_three(candidate):
    return detect_number(reduced_three, candidate)

def detect_four(candidate):
    return detect_number(reduced_four, candidate)

def detect_five(candidate):
    return detect_number(reduced_five, candidate)

def detect_six(candidate):
    return detect_number(reduced_six, candidate)

def detect_seven(candidate):
    return detect_number(reduced_seven, candidate)

def detect_eight(candidate):
    return detect_number(reduced_eight, candidate)

def detect_nine(candidate):
    return detect_number(reduced_nine, candidate)

def detect_zero(candidate):
    return detect_number(reduced_zero, candidate)

detect_fcts = [detect_zero, detect_one, detect_two, detect_three, detect_four, 
               detect_five, detect_six, detect_seven, detect_eight, detect_nine]

def recognize_digit(candidate):
    convs = [dfct(candidate) for dfct in detect_fcts]
    print(candidate.shape, convs)
    return convs.index(max(convs))
```

```python
im = cv2.imread("./fapiao/croped/80.14.png")
imgray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(imgray, 145, 255, 0)
plot_img('thresh', thresh)
height_of_digit = 20
width_of_digit = 13
#im2, contours = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
for cnt in contours:
    [x, y, w, h] = cv2.boundingRect(cnt)
    if (w < width_of_digit * 0.7 or h < height_of_digit * 0.7):
        continue
    if (w > width_of_digit * 2 or h > height_of_digit * 2):
        continue
    cv2.rectangle(thresh, (x - 1, y - 1), (x + 1 + w, y + 1 + h), (0, 0, 255), 1)
    roi = thresh[y:y + h, x:x + w]
    roi = cv2.adaptiveThreshold(roi,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV,21,11)
    cv2.imshow('Training: Enter digits displayed in the red rectangle!', roi)
    try:
        print("Digit is:", recognize_digit(roi))
    except Exception as e:
        print("Size Exception", e)
        cv2.destroyAllWindows()
        break
    cv2.waitKey(0)
cv2.destroyAllWindows()
```
